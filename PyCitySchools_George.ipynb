```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#list out all the names of the schools
schoolnames = school_data_complete["school_name"].unique()
#count the length of the school name list 
schoolcount = len(school_data["school_name"])
#count the number of students by counting the individual student names with count 
totalstudents = school_data_complete["student_name"].count()
#total school budget summed from school_data sheet 
totalbudget = school_data["budget"].sum()
#average the math score data set 
averagemath = school_data_complete["math_score"].mean()
#average the reading score data set 
averagereading = school_data_complete["reading_score"].mean()
#count the number of students that had a passing math score 
mathpass = school_data_complete.loc[school_data_complete["math_score"] >= 70,["Student ID"]].count()
#get percentage by count over total students
mathpassstudents= mathpass/totalstudents
#count the number of students that had a passing reading score
readingpass = school_data_complete.loc[school_data_complete["reading_score"] >= 70,["Student ID"]].count()
#get percentage by count over total students 
readingpassstudents = readingpass/totalstudents
#count the number of students that had both passing math and passing reading scores 
readingmath = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
#get percentage by count over total students 
readingmathscores = len(readingmath)/totalstudents 

#Create a dataframe 
schoolsummary = pd.DataFrame({
    "Total Schools": schoolcount,
    "Total Students": totalstudents,
    "Total School Budget": totalbudget,
    "Average Math Score": averagemath,
    "Average Reading Score": averagereading,
    "% Passing Math": mathpassstudents,
    "% Passing Reading": readingpassstudents,
    "Overall Passing Rate": readingmathscores})

#format dataframe 
schoolsummary["Total Students"] = schoolsummary["Total Students"].astype(float).map("{:,.0f}".format)
schoolsummary["Total School Budget"] = schoolsummary["Total School Budget"].astype(float).map("${:,.2f}".format)
schoolsummary["Average Math Score"] = schoolsummary["Average Math Score"].astype(float).map("{:,.2f}".format)
schoolsummary["Average Reading Score"] = schoolsummary["Average Reading Score"].astype(float).map("{:,.2f}".format)
schoolsummary["% Passing Math"] = schoolsummary["% Passing Math"].astype(float).map("{:.2%}".format)
schoolsummary["% Passing Reading"] = schoolsummary["% Passing Reading"].astype(float).map("{:.2%}".format)
schoolsummary["Overall Passing Rate"] = schoolsummary["Overall Passing Rate"].astype(float).map("{:.2%}".format)

schoolsummary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schoolgroup = school_data_complete.groupby(["school_name"])
schools = school_data["school_name"].tolist()
schooltypes = school_data['type'].tolist()
students = schoolgroup['Student ID'].count().tolist()
budget = school_data["budget"].tolist()
groupschool = school_data_complete.groupby("school_name")
perstudent = school_data["budget"]/school_data["size"]
math = round(groupschool["math_score"].mean(),2).tolist()
reading = round(groupschool["reading_score"].mean(),2).tolist()
mathpassedpercent = round(school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/students,4).tolist()
readingpassedpercent = round((school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/students),4).tolist()
passboth = school_data_complete[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70)]
schoolgroupbyboth = passboth.groupby(["school_name"])
passbothcount = schoolgroupbyboth["School ID"].count()
bothpassedpercent = round(school_data_complete[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70)].groupby("school_name")["Student ID"].count()/students,4).tolist()

#Make the dataframe 
eachschoolsummary = pd.DataFrame({
    "School":schools,
    "School Type":schooltypes,
    "Total Students": students, 
    "Total Budget":budget,
    "Per Student Budget":perstudent, 
    "Average Math Score":math, 
    "Average Reading Score":reading,
    "% Passing Math":mathpassedpercent,
    "% Passing Reading": readingpassedpercent,
    "Overall Passing Rate": bothpassedpercent})

#Format the table 
eachschoolsummary["Total Students"] = eachschoolsummary["Total Students"].astype(float).map("{:,}".format)
eachschoolsummary["Total Budget"] = eachschoolsummary["Total Budget"].astype(float).map("${:,}".format)
eachschoolsummary["Per Student Budget"] = eachschoolsummary["Per Student Budget"].astype(float).map("${:,}".format)
eachschoolsummary["Average Math Score"] = eachschoolsummary["Average Math Score"].astype(float).map("{:,}".format)
eachschoolsummary["Average Reading Score"] = eachschoolsummary["Average Reading Score"].astype(float).map("{:,}".format)
eachschoolsummary["% Passing Math"] = eachschoolsummary["% Passing Math"].astype(float).map("{:.2%}".format)
eachschoolsummary["% Passing Reading"] = eachschoolsummary["% Passing Reading"].astype(float).map("{:.2%}".format)
eachschoolsummary["Overall Passing Rate"] = eachschoolsummary["Overall Passing Rate"].astype(float).map("{:.2%}".format)

eachschoolsummary.set_index("School")
```

## Math Scores by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# create spending bins
bins = [0, 584, 614, 644, 680]
spendbinlabels = ["< $585", "$585 - 614", "$615 - 644", "$645 - 675"]
school_data_complete["spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels = spendbinlabels)

#group by spending
spendgroups = school_data_complete.groupby("spending_bins")

#calculations
mathscore = spendgroups["math_score"].mean()
readingscore = spendgroups["reading_score"].mean()
percentmathpass = school_data_complete[school_data_complete["math_score"] >= 70].groupby("spending_bins")["Student ID"].count()/spendgroups["Student ID"].count()
percentreadingpass = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("spending_bins")["Student ID"].count()/spendgroups["Student ID"].count()
percentbothpass = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("spending_bins")["Student ID"].count()/spendgroups["Student ID"].count()
            
#create a dataframe 
scoresbyspend = pd.DataFrame({"Average Math Score": mathscore,"Average Reading Score": readingscore,"% Passing Math": percentmathpass,"% Passing Reading": percentreadingpass, "Overall Passing Rate": percentbothpass})

#Format list 
scoresbyspend["Average Math Score"] = scoresbyspend["Average Math Score"].astype(float).map("{:,.2f}".format)
scoresbyspend["Average Reading Score"] = scoresbyspend["Average Reading Score"].astype(float).map("{:,.2f}".format)
scoresbyspend["% Passing Math"] = scoresbyspend["% Passing Math"].astype(float).map("{:.2%}".format)
scoresbyspend["% Passing Reading"] = scoresbyspend["% Passing Reading"].astype(float).map("{:.2%}".format)
scoresbyspend["Overall Passing Rate"] = scoresbyspend["Overall Passing Rate"].astype(float).map("{:.2%}".format)          
scoresbyspend.index.name = "Per Student Budget"
scoresbyspend
```

## Scores by School Size

```python
# create population bins
bins = [0, 999, 1999, 5001]
populationlabels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["sizebins"] = pd.cut(school_data_complete["size"], bins, labels = populationlabels)

#group by size of population
groupbysize = school_data_complete.groupby("sizebins")
groupbysize.head()

#calculations 
mathaverage = groupbysize["math_score"].mean()
readingaverage = groupbysize["reading_score"].mean()
percentmath = school_data_complete[school_data_complete["math_score"] >= 70].groupby("sizebins")["size"].count()/groupbysize["size"].count()
percentreading = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("sizebins")["size"].count()/groupbysize["size"].count()
overallpassing = school_data_complete[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70)].groupby("sizebins")["size"].count()/groupbysize["sizebins"].count()
            
#Build a dataframe        
scoresgroupbysize = pd.DataFrame({"Average Math Score": mathaverage,"Average Reading Score": readingaverage,"% Passing Math": percentmath,"% Passing Reading": percentreading, "Overall Passing Rate": overallpassing})

#format the dataframe 
scoresgroupbysize["Average Math Score"] = scoresgroupbysize["Average Math Score"].astype(float).map("{:,.2f}".format)
scoresgroupbysize["Average Reading Score"] = scoresgroupbysize["Average Reading Score"].astype(float).map("{:,.2f}".format)
scoresgroupbysize["% Passing Math"] = scoresgroupbysize["% Passing Math"].astype(float).map("{:.2%}".format)
scoresgroupbysize["% Passing Reading"] = scoresgroupbysize["% Passing Reading"].astype(float).map("{:.2%}".format)
scoresgroupbysize["Overall Passing Rate"] = scoresgroupbysize["Overall Passing Rate"].astype(float).map("{:.2%}".format)
scoresgroupbysize.index.name = "Per Size of School"
scoresgroupbysize
```
